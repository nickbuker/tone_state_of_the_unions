```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from src import plotting
%matplotlib inline

from pandas.tools.plotting import scatter_matrix
```

## Load tones by year json in dataframe

```python
df = pd.read_json('./data/tones_year.json')
df = df.T
df.head()
```

## Plot individual tones

```python
# create colors dict
colors = {'Anger': 'r', 'Sadness': 'b', 'Joy':'g',
          'Fear': 'm', 'Disgust': 'c'}
```

```python
plotting.individual_tones(df = df,
                          tone = 'Anger',
                          colors = colors,
                          title = 'Anger',
                          y_lab = 'Anger Score')
plt.savefig('./images/py_anger.png')
```

```python
plotting.individual_tones(df = df,
                          tone = 'Sadness',
                          colors = colors,
                          title = 'Sadness',
                          y_lab = 'Sadness Score')
plt.savefig('./images/py_sadness.png')
```

```python
plotting.individual_tones(df = df,
                          tone = 'Joy',
                          colors = colors,
                          title = 'Joy',
                          y_lab = 'Joy Score')
plt.savefig('./images/py_joy.png')
```

```python
plotting.individual_tones(df = df,
                          tone = 'Fear',
                          colors = colors,
                          title = 'Fear',
                          y_lab = 'Fear Score')
plt.savefig('./images/py_fear.png')
```

```python
plotting.individual_tones(df = df, 
                          tone = 'Disgust',
                          colors = colors,
                          title = 'Disgust',
                          y_lab = 'Disgust Score')
plt.savefig('./images/py_disgust.png')
```

## Plot multiple tones 

```python
tone_labels = ['Anger', 'Sadness', 'Joy', 'Fear', 'Disgust']
plotting.multiple_tones(df = df,
                        tone_labels = tone_labels,
                        colors = colors,
                        title = 'All Tones',
                        figsize=(12, 12),
                        y_lim=(0, 0.75))
plt.savefig('./images/py_all_tones.png')
```

```python
tone_labels = ['Anger', 'Fear']
plotting.multiple_tones(df = df,
                        tone_labels = tone_labels,
                        colors = colors,
                        title = 'Fear and Anger')
plt.savefig('./images/py_fear_anger.png')
```

```python
tone_labels = ['Joy', 'Sadness']
plotting.multiple_tones(df = df,
                        tone_labels = tone_labels,
                        colors = colors,
                        title = 'Joy and Sadness')
```

```python
tone_labels = ['Joy', 'Anger']
plotting.multiple_tones(df = df,
                        tone_labels = tone_labels,
                        colors = colors,
                        title = 'Joy and Anger')
```
