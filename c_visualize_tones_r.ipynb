```r
library(ggplot2)
```

## Load data by year csv in data frame

```r
df <- read.csv('./data/tones_year.csv')
```

```r
head(df, 5)
```

## Plot individual tones

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Anger), color = 'red') +
geom_point(aes(x = X, y = Anger), color = 'red') +
geom_hline(yintercept = mean(df$Anger), lty = 2) +
ggtitle('Anger')
```

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Sadness), color = 'blue') +
geom_point(aes(x = X, y = Sadness), color = 'blue') +
geom_hline(yintercept = mean(df$Sadness), lty = 2) +
ggtitle('Sadness')
```

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Joy), color = 'green3') +
geom_point(aes(x = X, y = Joy), color = 'green3') +
geom_hline(yintercept = mean(df$Joy), lty = 2) +
ggtitle('Joy')
```

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Fear), color = 'purple') +
geom_point(aes(x = X, y = Fear), color = 'purple') +
geom_hline(yintercept = mean(df$Fear), lty = 2) +
ggtitle('Fear')
```

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Disgust), color = 'cyan2') +
geom_point(aes(x = X, y = Disgust), color = 'cyan2') +
geom_hline(yintercept = mean(df$Disgust), lty = 2) +
ggtitle('Disgust')
```

## Plot multiple tones

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Anger), color = 'red') +
geom_point(aes(x = X, y = Anger), color = 'red') +
geom_line(aes(x = X, y = Sadness), color = 'blue') +
geom_point(aes(x = X, y = Sadness), color = 'blue') +
geom_line(aes(x = X, y = Joy), color = 'green3') +
geom_point(aes(x = X, y = Joy), color = 'green3') +
geom_line(aes(x = X, y = Fear), color = 'purple') +
geom_point(aes(x = X, y = Fear), color = 'purple') +
geom_line(aes(x = X, y = Disgust), color = 'cyan2') +
geom_point(aes(x = X, y = Disgust), color = 'cyan2') +
ggtitle('All Tones')
```

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Anger), color = 'red') +
geom_point(aes(x = X, y = Anger), color = 'red') +
geom_line(aes(x = X, y = Fear), color = 'purple') +
geom_point(aes(x = X, y = Fear), color = 'purple') +
labs(title='Fear and Anger', y='Score')
```

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Sadness), color = 'blue') +
geom_point(aes(x = X, y = Sadness), color = 'blue') +
geom_line(aes(x = X, y = Joy), color = 'green3') +
geom_point(aes(x = X, y = Joy), color = 'green3') +
labs(title='Joy and Sadness', y='Score')
```

```r
ggplot(data = df) + 
geom_line(aes(x = X, y = Anger), color = 'red') +
geom_point(aes(x = X, y = Anger), color = 'red') +
geom_line(aes(x = X, y = Joy), color = 'green3') +
geom_point(aes(x = X, y = Joy), color = 'green3') +
labs(title='Joy and Anger', y='Score')
```
