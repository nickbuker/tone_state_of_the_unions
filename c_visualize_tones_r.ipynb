```r
library(ggplot2)
```

## Load data by year csv in data frame

```r
df <- read.csv('./data/tones_year.csv')
```

```r
head(df, 5)
```

## Plot individual tones

```r
p = qplot(df$X, df$Anger, xlab ='Year', ylab = 'Anger Score') + geom_point(color='red') + geom_line(color='red') 
h = geom_hline(yintercept = mean(df$Anger), lty = 2)
p + h
```

```r
p = qplot(df$X, df$Sadness, xlab ='Year', ylab = 'Sadness Score') + geom_point(color='blue') + geom_line(color='blue') 
h = geom_hline(yintercept = mean(df$Sadness), lty = 2)
p + h
```

```r
p = qplot(df$X, df$Joy, xlab ='Year', ylab = 'Joy Score') + geom_point(color='green3') + geom_line(color='green3') 
h = geom_hline(yintercept = mean(df$Joy), lty = 2)
p + h
```

```r
p = qplot(df$X, df$Fear, xlab ='Year', ylab = 'Fear Score') + geom_point(color='purple') + geom_line(color='purple') 
h = geom_hline(yintercept = mean(df$Fear), lty = 2)
p + h
```

```r
p = qplot(df$X, df$Disgust, xlab ='Year', ylab = 'Disgust Score') + geom_point(color='cyan2') + geom_line(color='cyan2') 
h = geom_hline(yintercept = mean(df$Disgust), lty = 2)
p + h
```
